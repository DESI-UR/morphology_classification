# Add Cutouts to SGA data table

```python
import os
import numpy as np
import fitsio
from astropy.io import fits
from astropy.table import Table
from tqdm import tqdm
import math
```

```python
# Function to get RA slice directory name
def get_raslice(ra):
    return f'{int(ra):03d}'
def addPath(df, ra, dec, id, g, z, r, morphology, path):
    length = len(df)
    processed_galaxies = []
    no_image_path = []
    resized = Table()
    ref_ids = resized["ref_id"] = df[id]
    ras = resized["ra"] = df[ra]
    decs = resized["dec"] = df[dec]
    resized['r_mag'] = df[r]
    resized['Morphology'] = df[morphology]
    paths = resized['Path'] = np.full(length, '', dtype=f'<U255')

# Read in the catalog
#sga2020 = Table.read('/global/cfs/cdirs/cosmo/data/sga/2020/SGA-2020.fits', hdu='ELLIPSE')
sgaBeta2025 = Table.read('/pscratch/sd/q/qshimp/SGA2025-data/sga2025_sizedTargets1000.csv')
length = len(sgaBeta2025)
# Lists to store issues encountered
processed_galaxies = []
no_image_path = []
image_shape_unknown = []

# Initialize Resized Table
SGA_Resized = Table()
#SGA_Resized["SGA_ID"] = sga2020['SGA_ID']
#SGA_Resized['RA'] = sga2020['RA']
#SGA_Resized['DEC'] = sga2020['DEC']
#SGA_Resized['G_MAG_SB26'] = sga2020['G_MAG_SB26']
#SGA_Resized['Z_MAG_SB26'] = sga2020['Z_MAG_SB26']
#SGA_Resized['R_MAG_SB26'] = sga2020['R_MAG_SB26']
#SGA_Resized['Morphology'] = sga2020['MORPHTYPE']
#SGA_Resized['Path'] = np.full(383620, '', dtype=f'<U255')

ref_ids = SGA_Resized["ref_id"] = sgaBeta2025['TargetID']
ras = SGA_Resized["ra"] = sgaBeta2025['Target_RA']
decs = SGA_Resized["dec"] = sgaBeta2025['Target_DEC']
SGA_Resized['r_mag'] = sgaBeta2025['mag']
SGA_Resized['Morphology'] = [''] * length
paths = SGA_Resized['Path'] = np.full(length, '', dtype=f'<U255')

# Process a range of galaxies
galaxy_range = range(length)

# Base directory for the FITS files
datadir = '/pscratch/sd/q/qshimp/Cutouts/sga2025/dr10/'

# Loop through each galaxy in the specified range
for i in tqdm(galaxy_range, desc="Processing Galaxies", unit="galaxy", leave=True):
    ref_id = ref_ids[i]
    ra = ras[i]
    dec = decs[i]

    # Generate the path based on the provided directory structure
    path = os.path.join(datadir, f'{ref_id}_grz_152_{ra:.4f}_{dec:.4f}.fits')
    
    if os.path.exists(path):
        fits_file = fits.open(path, ignore_missing_simple=True, ignore_missing_end=True)
        image = fits_file[0].data
        fits_file.close()

        #print("image shape:", image.shape)

        # append filenames to new Resized Path Col in SGA_Resized Table
        paths[i] = path

    else:
        no_image_path.append(path)

SGA_Resized['Path'] = paths

# Save the processed data to a new FITS file
SGA_Resized.write("/pscratch/sd/q/qshimp/Sorter/152by152_sga2025_1000test.fits", overwrite=True)
```

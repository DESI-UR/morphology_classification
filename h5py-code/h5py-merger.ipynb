```python
import h5py
import numpy as np
import glob

output_file = '/pscratch/sd/j/jlargett/DESI_SGA_MINE/Sorter/h5py-files/SGA2020.h5'
file_list = glob.glob('/pscratch/sd/j/jlargett/DESI_SGA_MINE/Sorter/h5py-files/bounded-h5py/SGA-152by152-Cutouts-*.h5')

with h5py.File(output_file, 'w') as f_out:
    first_file = True
    for file in file_list:
        with h5py.File(file, 'r') as f_in:
            if first_file:
                # Initialize datasets in the output file
                for key in f_in.keys():
                    data_shape = (0,) + f_in[key].shape[1:]
                    max_shape = (None,) + f_in[key].shape[1:]
                    f_out.create_dataset(key, shape=data_shape, maxshape=max_shape, chunks=True, dtype=f_in[key].dtype)
                first_file = False
            
            # Append data from each file
            for key in f_in.keys():
                f_out[key].resize((f_out[key].shape[0] + f_in[key].shape[0]), axis=0)
                f_out[key][-f_in[key].shape[0]:] = f_in[key][:]
```

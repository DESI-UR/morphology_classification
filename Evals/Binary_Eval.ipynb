### Binary Evaluation Historgram

```python
# Define a function to classify the number of agreements
def classify_agreements(row):
    # Extract predictions
    predictions = [row['Spiral_Pred_Lab'], row['Elliptical_Pred_Lab'], row['Lenticular_Pred_Lab'], row['Irregular_Pred_Lab']]
    
    # Count 'Other' and non-'Other' predictions
    count_other = predictions.count('Other')
    count_non_other = 4 - count_other  # Total number of categories

    # Determine the type of agreement
    if count_other == 0:
        return 'Crying (0 Others)'
    elif count_other == 1:
        return 'Weak Partial Agreement (1 Other)'
    elif count_other == 2:
        return 'Strong Partial Agreement (2 Others)'
    elif count_other == 3:
        return 'Total Agreement (3 Others)'
    elif count_other == 4:
        return 'No Classification (4 Others)'

# Apply the classification function to each row
merged_df_FP['Agreement_Type'] = merged_df_FP.apply(classify_agreements, axis=1)

# Count the number of rows for each agreement type
agreement_counts = merged_df_FP['Agreement_Type'].value_counts().sort_index()

# Print the counts of each agreement type
print("Agreement Type Counts:")
print(agreement_counts)
```

```python
# Define the order of agreement types
order = [
    'Crying (0 Others)',
    'Weak Partial Agreement (1 Other)',
    'Strong Partial Agreement (2 Others)',
    'Total Agreement (3 Others)',
    'No Classification (4 Others)',
]

# Ensure the DataFrame has the categories from least amount of agreement to most
merged_df_FP['Agreement_Type'] = pd.Categorical(merged_df_FP['Agreement_Type'], categories=order, ordered=True)
agreement_counts = merged_df_FP['Agreement_Type'].value_counts().reindex(order)

# Define a medium pink color :)
medium_pink = '#f07a9c'
# Make histogram bars pink
bar_colors = [medium_pink] * len(order)

# Plot the counts of each agreement type
plt.figure(figsize=(12, 7))
bars = agreement_counts.plot(kind='bar', color=bar_colors, edgecolor='black')
plt.title('Counts of Each Agreement Type', fontsize=16)
plt.xlabel('Agreement Type', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add counts above each bar
for bar in bars.patches:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height + 10, int(height), ha='center', va='bottom', fontsize=12)

# Display Plot
plt.tight_layout()
plt.show()
```

```python
# define a func to remove rows where the SGA_ID column contains any value in exclude_ids
def filter_sga_ids(base_df, exclude_ids):
    return base_df[~base_df['SGA_ID'].isin(exclude_ids)]

# Process each galaxy type separately
def process_galaxy_type(it1_dfs, it2_df, it3_df):
    # Extract SGA IDs to exclude
    ids_in_it3 = set(it3_df['SGA_ID'])
    ids_in_it2 = set(it2_df['SGA_ID'])
    # Collect all SGA_IDs from it3_df (Iteration 3) and it2_df (Iteration 2) into sets (ids_in_it3, ids_in_it2).

    # Remove Iteration 3 SGA IDs from Iterations 1 and 2
    filtered_it1 = [filter_sga_ids(df, ids_in_it3) for df in it1_dfs]
    # For each DataFrame in it1_dfs, remove rows where SGA_ID is in ids_in_it3 using the filter_sga_ids function.
    filtered_it2 = filter_sga_ids(it2_df, ids_in_it3)
    # Remove rows from it2_df where SGA_ID is in ids_in_it3.

    # Remove Iteration 2 SGA IDs from Iteration 1
    filtered_it1 = [filter_sga_ids(df, ids_in_it2) for df in filtered_it1]
    # Iteration 1 is further filtered to remove rows where SGA_ID is in ids_in_it2.
    
    # Combine the filtered DataFrames
    combined_df = pd.concat(filtered_it1 + [filtered_it2, it3_df])
    # The filtered list of Iteration 1 DataFrames (filtered_it1)., The filtered Iteration 2 DataFrame (filtered_it2).The unfiltered Iteration 3 DataFrame (it3_df).
    return combined_df
```

```python
import pandas as pd
from astropy.table import Table
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display
from collections import defaultdict
import os
```

```python
it1 = Table.read("/pscratch/sd/j/jlargett/DESI_SGA_MINE/Sorter/fpy3_elliptical_galaxies_it1.csv")
```

```python
it2 = Table.read("/pscratch/sd/j/jlargett/DESI_SGA_MINE/Sorter/fpy3_elliptical_galaxies_it2.csv")
```

```python
gal_dataset = Table.read('/pscratch/sd/j/jlargett/DESI_SGA_MINE/Sorter/FP/FP_Targets_SSL_wanchors.csv')
```

```python
# Get targetids with Predicted_Label == 1
remove_ids = set(it1["Target_ID"][it1["Predicted_Label"] == 1])

# Keep rows whose targetid is NOT in remove_ids
mask = np.array([tid not in remove_ids for tid in gal_dataset["targetid"]])

gal_dataset_filtered = gal_dataset[mask]

print(f"Original size: {len(gal_dataset):,}")
print(f"Removed: {len(gal_dataset) - len(gal_dataset_filtered):,}")
print(f"Remaining: {len(gal_dataset_filtered):,}")
```

```python
# Save
gal_dataset_filtered.write(
    "/pscratch/sd/j/jlargett/DESI_SGA_MINE/Sorter/FP/FP_Targets_SSL_wanchors_it2.csv",
    overwrite=True
)
```

```python
class_lookup = {
    int(row["Target_ID"]): str(row["Predicted_Type"])
    for row in it1
}

# it2 overwrites it1 for overlapping targetids
for row in it2:
    class_lookup[int(row["Target_ID"])] = str(row["Predicted_Type"])

# Add classification column to gal_dataset
classification = np.array([
    class_lookup.get(int(tid), "Unclassified")
    for tid in gal_dataset["targetid"]
])

gal_dataset["classification"] = classification

# Check counts
unique, counts = np.unique(gal_dataset["classification"], return_counts=True)
for u, c in zip(unique, counts):
    print(f"{u}: {c}")
```
